# Fibers length in multi animals
This example uses synthetic data to showcase how `cuisto` can be used in a [pipeline](../guide-pipeline.html).

Annotations measurements should be exported from QuPath, following the required [directory structure](../guide-pipeline.html#directory-structure).

Alternatively, you can merge all your CSV files yourself, one per animal, adding an animal ID to each table. Those can be processed with the `cuisto.process.process_animal()` function, in a loop, collecting the results at each iteration and finally concatenating the results. Finally, those can be used with `display` module. See the API reference for the [`process` module](../api-process.html) and the related example [`batch_process_animals.py`](https://github.com/TeamNCMC/cuisto/blob/main/examples/batch_process_animals.py).

```python
import cuisto
```

```python
# Full path to your configuration file, edited according to your need beforehand
config_file = "../../resources/demo_config_multi.toml"
```

```python
# Files
wdir = "../../resources/multi"
animals = ["mouse0", "mouse1"]
```

```python
# get configuration
cfg = cuisto.Config(config_file)
```

```python
# get distributions per regions
df_regions, _, _ = cuisto.process.process_animals(
    wdir, animals, cfg, compute_distributions=False
)

# have a look
display(df_regions.head(10))
```

```python
figs_regions = cuisto.display.plot_regions(df_regions, cfg)
```
